#Numpy

1. Tipo de Dado:

**Listas:** podem armazenar elementos de diferentes tipos de dados, como números, strings, booleanos e até mesmo outras listas. Isso as torna versáteis para tarefas gerais.

**Arrays NumPy:** são homogêneos, ou seja, todos os elementos devem ser do mesmo tipo de dado, como int, float ou bool. Isso os torna mais eficientes para cálculos matemáticos.


2. Funcionalidade:

**Listas:** oferecem funções integradas para tarefas como inserção, remoção, ordenação e fatiamento. Elas são flexíveis e fáceis de usar para manipulação geral de dados.

**Arrays NumPy:** fornecem funções avançadas para operações matemáticas, como álgebra linear, estatística e processamento de sinal. Elas são mais adequadas para análise e processamento de dados numéricos.

# O que é e como é um DataFrame?

Um DataFrame é uma estrutura de dados bidimensional do Pandas, que se assemelha a uma tabela em uma base de dados ou a uma planilha do Excel. Ele possui colunas e linhas rotuladas, onde cada coluna pode conter um tipo de dado diferente (números, strings, datas, etc.).

###Características do DataFrame:
**Etiquetas (Rótulos):** Tanto as linhas quanto as colunas possuem rótulos. As colunas são acessadas através de seus nomes e as linhas geralmente são acessadas por índices numéricos.

**Heterogeneidade:** Diferentes colunas podem conter diferentes tipos de dados (inteiros, floats, strings, etc.).

**Alinhamento Automático:**  Operações entre DataFrames automaticamente alinham os dados baseados nos rótulos das linhas e colunas.

*Exemplo de Criação de um DataFrame:*
Aqui está um exemplo de como criar um DataFrame a partir de um dicionário:
```
import pandas as pd

# Dados para o DataFrame
data = {
    'Nome': ['Ana', 'João', 'Maria'],
    'Idade': [28, 34, 29],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte']
}

# Criação do DataFrame
df = pd.DataFrame(data)

# Exibição do DataFrame
print(df)
```

````python
# definindo uma função para 'limpar' dados

def limpa_preco(linha):
  linha['price'] = linha ['price'].replace('R$', '')
  linha['price'] = linha['price'].replace('\n', '')
  linha['price'] = linha['price'].replace('/Mês', '')
  linha['price'] = linha['price'].replace('/Ano', '')
  linha['price'] = linha['price'].replace('/Dia', '')
  linha['price'] = linha['price'].replace('                     ', '')
  linha['price'] = linha['price'].replace(' ', '')
  return linha
````

````python
limpa_preco(auxiliar)
#auxiliar
````

````python
df.apply(lambda x: limpa_preco(x), axis=1)
````

````python
def ajusta_alugueis(linha):
  linha['price'] = str(linha['price']).replace('.', '')
  preco = int(linha['price'])
  if(preco < 10000):
    preco = preco * 200
  linha['price'] = preco
  return linha
````

## **Matplotlib**

A biblioteca Matplotlib é uma das bibliotecas mais populares e amplamente usadas em Python para visualização de dados. Ela permite a criação de uma ampla variedade de gráficos e visualizações, desde gráficos de linha simples até visualizações complexas em 2D e 3D. A principal função de Matplotlib é ajudar os usuários a entender e interpretar seus dados através de gráficos visuais.

1. **`plt.plot()`**:
   - **Uso**: Criação de gráficos de linha.
   - **Exemplo**: `plt.plot(x, y)`

2. **`plt.scatter()`**:
   - **Uso**: Criação de gráficos de dispersão (scatter plots).
   - **Exemplo**: `plt.scatter(x, y)`

3. **`plt.bar()`**:
   - **Uso**: Criação de gráficos de barras.
   - **Exemplo**: `plt.bar(x, height)`

4. **`plt.hist()`**:
   - **Uso**: Criação de histogramas.
   - **Exemplo**: `plt.hist(data)`

5. **`plt.pie()`**:
   - **Uso**: Criação de gráficos de pizza.
   - **Exemplo**: `plt.pie(sizes, labels=labels)`

6. **`plt.xlabel()`**:
   - **Uso**: Definição de rótulo para o eixo x.
   - **Exemplo**: `plt.xlabel('Eixo X')`

7. **`plt.ylabel()`**:
   - **Uso**: Definição de rótulo para o eixo y.
   - **Exemplo**: `plt.ylabel('Eixo Y')`

8. **`plt.title()`**:
   - **Uso**: Definição de título para o gráfico.
   - **Exemplo**: `plt.title('Título do Gráfico')`

9. **`plt.legend()`**:
   - **Uso**: Adição de uma legenda ao gráfico.
   - **Exemplo**: `plt.legend(['Linha 1', 'Linha 2'])`

10. **`plt.grid()`**:
    - **Uso**: Adição de uma grade ao gráfico.
    - **Exemplo**: `plt.grid(True)`

11. **`plt.figure()`**:
    - **Uso**: Criação de uma nova figura.
    - **Exemplo**: `plt.figure()`

12. **`plt.subplot()`**:
    - **Uso**: Criação de subplots (vários gráficos em uma única figura).
    - **Exemplo**: `plt.subplot(1, 2, 1)`

13. **`plt.show()`**:
    - **Uso**: Exibição do gráfico.
    - **Exemplo**: `plt.show()`

14. **`plt.savefig()`**:
    - **Uso**: Salvamento do gráfico como um arquivo de imagem.
    - **Exemplo**: `plt.savefig('grafico.png')`

A função `sns.pairplot()` do Seaborn é uma ferramenta poderosa para explorar visualmente as relações entre variáveis em um DataFrame. Ela cria uma grade de gráficos que mostra gráficos de dispersão (scatter plots) para cada par de variáveis numéricas, além de histogramas ou gráficos de densidade para a distribuição univariada de cada variável ao longo da diagonal da grade.

## O Que pairplot Faz
Visualização de Relações Bivariadas:

Para cada par de variáveis numéricas no DataFrame, pairplot cria um gráfico de dispersão mostrando como essas variáveis se relacionam. Isso ajuda a identificar correlações e padrões entre variáveis.
Distribuição Univariada:

Nos gráficos ao longo da diagonal da grade, pairplot mostra a distribuição de cada variável. Normalmente, isso é feito usando histogramas ou gráficos de densidade (kernel density plots).
Facetação por Categorias:

pairplot permite dividir os dados em categorias e usar diferentes cores para diferentes categorias, facilitando a visualização de como diferentes grupos de dados se comportam.
##Principais Argumentos
`data`: O DataFrame que contém os dados a serem visualizados.
`hue`: Nome da coluna que contém categorias para colorir os pontos no gráfico, facilitando a distinção entre diferentes grupos.
`vars`: Lista de variáveis para incluir na grade. Se não especificado, todas as variáveis numéricas serão usadas.
`kind`: Tipo de gráfico a ser usado para a distribuição univariada (pode ser 'scatter' para gráficos de dispersão ou 'reg' para incluir uma linha de regressão).
`diag_kind`: Tipo de gráfico a ser usado na diagonal (pode ser 'hist' para histogramas ou 'kde' para gráficos de densidade).

````python
import pandas as pd
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
````

Esses são importações de módulos da biblioteca scikit-learn, que é uma das bibliotecas mais populares em Python para aprendizado de máquina:

1. from sklearn.model_selection import train_test_split
train_test_split:

**O que é:** Uma função usada para dividir um conjunto de dados em dois subconjuntos: um conjunto de treinamento e um conjunto de teste.

**Uso:** Essa função é fundamental para validar modelos de aprendizado de máquina, pois permite avaliar o desempenho do modelo em dados não vistos (conjunto de teste).

2. from sklearn.linear_model import LogisticRegression
LogisticRegression:

**O que é:** Uma classe que implementa a regressão logística, um algoritmo de classificação usado para prever a probabilidade de uma amostra pertencer a uma das classes.

**Uso:** É amplamente usada para problemas de classificação binária (mas também pode ser usada para classificação multiclasse).

3. from sklearn.metrics import mean_squared_error
mean_squared_error:

**O que é:** Uma função usada para calcular o erro quadrático médio (MSE) entre valores reais e previstos.

**Uso:** MSE é uma métrica de avaliação comum para modelos de regressão. Mede a média dos quadrados das diferenças entre os valores reais e os valores previstos.

````python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from mlxtend.plotting import plot_confusion_matrix
from sklearn.metrics import confusion_matrix, classification_report
````

````python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/datasets/breast-cancer.csv')
df.head()
df.info()
````

````python
x = df.drop(['id', 'diagnosis'], axis=1)
x
````

````python
y = df['diagnosis']
y
````

````python
modelo = LogisticRegression(max_iter=10000)
````

````python
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y)
````

````python
modelo.fit(x_treino, y_treino)
````

````python
y_previsto = modelo.predict(x_teste)
````

````python
cm = confusion_matrix(y_previsto,y_teste)
````

````python
plot_confusion_matrix(cm)
````

````python
print(classification_report(y_previsto, y_teste))
````
