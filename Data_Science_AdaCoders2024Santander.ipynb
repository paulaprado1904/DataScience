#Numpy

1. Tipo de Dado:

**Listas:** podem armazenar elementos de diferentes tipos de dados, como números, strings, booleanos e até mesmo outras listas. Isso as torna versáteis para tarefas gerais.

**Arrays NumPy:** são homogêneos, ou seja, todos os elementos devem ser do mesmo tipo de dado, como int, float ou bool. Isso os torna mais eficientes para cálculos matemáticos.


2. Funcionalidade:

**Listas:** oferecem funções integradas para tarefas como inserção, remoção, ordenação e fatiamento. Elas são flexíveis e fáceis de usar para manipulação geral de dados.

**Arrays NumPy:** fornecem funções avançadas para operações matemáticas, como álgebra linear, estatística e processamento de sinal. Elas são mais adequadas para análise e processamento de dados numéricos.

#Principais Funções Usadas no Pandas:
O Pandas oferece uma vasta gama de funções para manipulação e análise de dados em tabelas (DataFrames). Abaixo, listo algumas das funções mais utilizadas, organizadas por categorias:

1. Leitura e Gravação de Dados:

- `pd.read_csv()`: Lê um arquivo CSV e o retorna como um DataFrame.
- `pd.read_excel()`: Lê um arquivo Excel e o retorna como um DataFrame.
- `pd.read_json()`: Lê um arquivo JSON e o retorna como um DataFrame.
- `pd.to_csv()`: Salva um DataFrame em um arquivo CSV.
- `pd.to_excel()`: Salva um DataFrame em um arquivo Excel.
- `pd.to_json()`: Salva um DataFrame em um arquivo JSON.

2. Seleção e Filtragem de Dados:

- `[]`: Seleciona colunas ou linhas por índice ou nome.
- `.loc()`: Seleciona elementos por rótulo de índice e/ou colunas.
- `.iloc()`: Seleciona elementos por posição de índice e/ou colunas.
- `.query()`: Seleciona elementos com base em uma expressão booleana.
- `.isin()`: Seleciona elementos que pertencem a um conjunto específico.

3. Manipulação de Dados:

- `.dropna()`: Remove linhas com valores ausentes (NaN).
- `.fillna()`: Preenche valores ausentes (NaN) com um valor específico.
- `.replace()`: Substitui valores em um DataFrame por outros valores.
- `.astype()`: Converte o tipo de dados de uma coluna.
- `.nunique()`: Retorna o número de valores únicos em uma coluna.
- `.apply()`: Aplica uma função a cada elemento de uma coluna ou DataFrame.

4. Agregação e Estatística Descritiva:

- `.groupby()`: Agrupa o DataFrame por uma ou mais colunas.
- `.agg()`: Aplica funções de agregação (soma, média, etc.) aos grupos.
- `.describe()`: Retorna estatísticas descritivas do DataFrame.
- `.head()`: Mostra as primeiras linhas do DataFrame.
- `.tail()`: Mostra as últimas linhas do DataFrame.

5. Junção e Combinação de Dados:

- `.merge()`: Combina dois DataFrames com base em uma coluna comum.
- `.concat()`: Concatena dois DataFrames verticalmente ou horizontalmente.
- `.append()`: Adiciona linhas de um DataFrame a outro.

6. Criação de Novas Colunas:

- `[]` com atribuição: Cria uma nova coluna com base em outra coluna ou expressão.
- `.assign()`: Cria uma ou mais novas colunas com base em funções ou valores.

# O que é e como é um DataFrame?

Um DataFrame é uma estrutura de dados bidimensional do Pandas, que se assemelha a uma tabela em uma base de dados ou a uma planilha do Excel. Ele possui colunas e linhas rotuladas, onde cada coluna pode conter um tipo de dado diferente (números, strings, datas, etc.).

###Características do DataFrame:
**Etiquetas (Rótulos):** Tanto as linhas quanto as colunas possuem rótulos. As colunas são acessadas através de seus nomes e as linhas geralmente são acessadas por índices numéricos.

**Heterogeneidade:** Diferentes colunas podem conter diferentes tipos de dados (inteiros, floats, strings, etc.).

**Alinhamento Automático:**  Operações entre DataFrames automaticamente alinham os dados baseados nos rótulos das linhas e colunas.

*Exemplo de Criação de um DataFrame:*
Aqui está um exemplo de como criar um DataFrame a partir de um dicionário:
```
import pandas as pd

# Dados para o DataFrame
data = {
    'Nome': ['Ana', 'João', 'Maria'],
    'Idade': [28, 34, 29],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte']
}

# Criação do DataFrame
df = pd.DataFrame(data)

# Exibição do DataFrame
print(df)
```

````python
# definindo uma função para 'limpar' dados

def limpa_preco(linha):
  linha['price'] = linha ['price'].replace('R$', '')
  linha['price'] = linha['price'].replace('\n', '')
  linha['price'] = linha['price'].replace('/Mês', '')
  linha['price'] = linha['price'].replace('/Ano', '')
  linha['price'] = linha['price'].replace('/Dia', '')
  linha['price'] = linha['price'].replace('                     ', '')
  linha['price'] = linha['price'].replace(' ', '')
  return linha
````

````python
limpa_preco(auxiliar)
#auxiliar
````

````python
df.apply(lambda x: limpa_preco(x), axis=1)
````

````python
def ajusta_alugueis(linha):
  linha['price'] = str(linha['price']).replace('.', '')
  preco = int(linha['price'])
  if(preco < 10000):
    preco = preco * 200
  linha['price'] = preco
  return linha
````

## **Matplotlib**

A biblioteca Matplotlib é uma das bibliotecas mais populares e amplamente usadas em Python para visualização de dados. Ela permite a criação de uma ampla variedade de gráficos e visualizações, desde gráficos de linha simples até visualizações complexas em 2D e 3D. A principal função de Matplotlib é ajudar os usuários a entender e interpretar seus dados através de gráficos visuais.

1. **`plt.plot()`**:
   - **Uso**: Criação de gráficos de linha.
   - **Exemplo**: `plt.plot(x, y)`

2. **`plt.scatter()`**:
   - **Uso**: Criação de gráficos de dispersão (scatter plots).
   - **Exemplo**: `plt.scatter(x, y)`

3. **`plt.bar()`**:
   - **Uso**: Criação de gráficos de barras.
   - **Exemplo**: `plt.bar(x, height)`

4. **`plt.hist()`**:
   - **Uso**: Criação de histogramas.
   - **Exemplo**: `plt.hist(data)`

5. **`plt.pie()`**:
   - **Uso**: Criação de gráficos de pizza.
   - **Exemplo**: `plt.pie(sizes, labels=labels)`

6. **`plt.xlabel()`**:
   - **Uso**: Definição de rótulo para o eixo x.
   - **Exemplo**: `plt.xlabel('Eixo X')`

7. **`plt.ylabel()`**:
   - **Uso**: Definição de rótulo para o eixo y.
   - **Exemplo**: `plt.ylabel('Eixo Y')`

8. **`plt.title()`**:
   - **Uso**: Definição de título para o gráfico.
   - **Exemplo**: `plt.title('Título do Gráfico')`

9. **`plt.legend()`**:
   - **Uso**: Adição de uma legenda ao gráfico.
   - **Exemplo**: `plt.legend(['Linha 1', 'Linha 2'])`

10. **`plt.grid()`**:
    - **Uso**: Adição de uma grade ao gráfico.
    - **Exemplo**: `plt.grid(True)`

11. **`plt.figure()`**:
    - **Uso**: Criação de uma nova figura.
    - **Exemplo**: `plt.figure()`

12. **`plt.subplot()`**:
    - **Uso**: Criação de subplots (vários gráficos em uma única figura).
    - **Exemplo**: `plt.subplot(1, 2, 1)`

13. **`plt.show()`**:
    - **Uso**: Exibição do gráfico.
    - **Exemplo**: `plt.show()`

14. **`plt.savefig()`**:
    - **Uso**: Salvamento do gráfico como um arquivo de imagem.
    - **Exemplo**: `plt.savefig('grafico.png')`


````python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/datasets/netflix.csv')
df = df.drop(df.columns[0], axis = 1)
# Remove a primeira coluna do DataFrame 'df'.
# A função 'drop' é usada para eliminar a coluna cujo nome é especificado em 'df.columns[0]',
# e 'axis=1' indica que estamos removendo uma coluna (não uma linha).
# O resultado é atribuído de volta a 'df', atualizando o DataFrame para não incluir a primeira coluna.

df.head()
````

````python
df['MAIN_GENRE'].value_counts()         #conta a quantidade de vezes que cada gênero ocorre
````

````python
df['MAIN_GENRE'].value_counts().keys()        # a função keys, retorna a lista dos valores de MIAN_GENRE

eixo_x = df['MAIN_GENRE'].value_counts().keys()  # Obtém os nomes dos gêneros únicos
eixo_y = df['MAIN_GENRE'].value_counts().values    # Obtém as contagens de cada gênero
````

````python
# Cria um gráfico de barras utilizando os dados fornecidos
# 'eixo_x' contém os nomes dos gêneros únicos e 'eixo_y' contém as contagens correspondentes
plt.bar(eixo_x, eixo_y)


#Titulo
plt.title('Melhores filmes por Gênero')

plt.xlabel('Generos de filme')
plt.ylabel('Quantidade de filmes')

plt.xticks(rotation=45)
# Exibe o gráfico na tela
plt.show()
````
